```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import plotly.graph_objs as go

import ssl
ssl._create_default_https_context = ssl._create_unverified_context
```

```python
tasa_paro = pd.read_csv('data/tasa_de_paro_en_comunidad_de_madrid_frente_al_conjunto_de_españa.csv')
paro_juvenil = pd.read_csv('data/evolucion_del_paro_juvenil_en_comunidad_de_madrid_y_el_conjunto_de_españa_.csv')
genero = pd.read_csv('data/genero_madrid.csv')
familias_en_paro = pd.read_csv('data/hogares_con_todos_sus_miembros_en_paro_en_comunidad_de_madrid_y_españa_.csv')
hogares = pd.read_csv('data/hogares.csv')
contratos = pd.read_csv('data/tipos_de_contratos_creados_en_comunidad_de_madrid.csv')
```

## Dataset del paro y su codigo

```python
tasa_paro
```

```python
f,ax1 = plt.subplots(figsize =(16,8))
z = tasa_paro['Año y Trimestre']
a = tasa_paro ['Comunidad de Madrid']
b = tasa_paro ['Tasa de paro en España']

plt.plot(z, a, 'o-', label='Comunidad de Madrid', color = 'black', alpha=0.8)
plt.plot(z, b,'o-',label='Tasa de paro en España',color = 'grey', alpha=0.8)

plt.xticks(rotation=45)
plt.xlabel('Periodo',fontsize = 15,color='black')
plt.ylabel('Tasas',fontsize = 15,color='black')
plt.title('Paro - Capital vs España',fontsize = 20,color='black')

plt.legend(loc='upper left')
plt.show()
```
